```python
# Replace missing values with NaN
time_df = time_df.replace('-', np.nan).astype({'spleen_times': 'float64', 'pancreas_times': 'float64'})

# Remove rows where both spleen_times and pancreas_times are NaN
filtered_df = time_df.dropna(subset=['spleen_times', 'pancreas_times'], how='all')

# Plot the data
plt.figure(figsize=(15, 5))

# Configure font properties for Times New Roman
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

# Plot spleen times
sns.scatterplot(x='amt_labels', y='spleen_times', data=filtered_df, label='Spleen', color='#701E83')

# Plot pancreas times
sns.scatterplot(x='amt_labels', y='pancreas_times', data=filtered_df, label='Pancreas', color='#EAAB3A')

# Add title and labels
plt.title('Labeling Time for Pancreas and Spleen')
plt.xlabel('Amount of Labels')
plt.ylabel('Time (min)')

# Set y-axis limits and ticks
plt.ylim(0, 40)
plt.yticks(range(0, 45, 5))

plt.savefig('/content/drive/MyDrive/spleen_pancreas_times_high_res.tiff', dpi=400)
# Show legend
plt.legend()

# Show plot
plt.show()
```
